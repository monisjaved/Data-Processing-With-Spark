```r
library(ggplot2)
```

```r
numfiles = c(2, 5, 10, 20, 30, 40, 50)
bigram = c(11.977, 12.312, 14.880, 17.033, 21.922, 25.227, 31.383)
trigram = c(13.405, 157.88, 161.27, 173.625, 522.699, 782.322, 1296.06)
data<-data.frame(numfiles, bigram, trigram)
```

```r
data
```

```r
ggplot(data=data, (aes(x = numfiles))) +
    geom_line(aes(y = bigram, colour = "bigram")) +
    geom_line(aes(y = trigram, colour = "trigram")) + 
    ylab(label="Time (s)") + 
    xlab("number of files run") +
    ggtitle("Comparison of scalabity of bigram and trigram on n files") +
    scale_colour_manual("", 
                      breaks = c( "bigram", "trigram"),
                      values = c( "green", "blue"))
```
